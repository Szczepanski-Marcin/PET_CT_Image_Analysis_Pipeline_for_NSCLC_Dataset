```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
import pandas as pd
import warnings
import SimpleITK as sitk
from lungmask import mask as lungmask

warnings.filterwarnings("ignore")
print("✅ Imports loaded")
```

```python
def normalize(img):
    return (img - img.min()) / (img.max() - img.min() + 1e-8)

def load_volume(dicom_dir):
    files = []
    for root, _, fnames in os.walk(dicom_dir):
        for f in fnames:
            if f.lower().endswith(".dcm"):
                files.append(os.path.join(root, f))

    slices = []
    for f in sorted(files):
        try:
            ds = pydicom.dcmread(f)
            img = ds.pixel_array.astype(float)
            slices.append(normalize(img))
        except:
            continue

    if len(slices) == 0:
        raise RuntimeError("❌ No DICOM slices loaded")

    return np.stack(slices, axis=0)
```

```python
def find_all_dicom_folders(root_path):
    dicom_folders = []
    for dirpath, _, files in os.walk(root_path):
        if any(f.lower().endswith(".dcm") for f in files):
            dicom_folders.append(dirpath)
    return dicom_folders

BASE_PATH = '/Users/Marcin_1/Desktop/PET/NSCLC/NSCLC Radiogenomics'
patient_id = "R01-001"
patient_path = os.path.join(BASE_PATH, patient_id)

folders = find_all_dicom_folders(patient_path)

if len(folders) == 0:
    raise RuntimeError("❌ No DICOM folders found")

print("Available scans:")
for i, f in enumerate(folders):
    print(i, f)

idx = int(input("Select scan index: "))
selected_scan = folders[idx]

volume = load_volume(selected_scan)
print("✅ Volume shape:", volume.shape)
```

```python
def lungmask_segment_safe(volume):
    img_sitk = sitk.GetImageFromArray(volume)
    inferer = lungmask.LMInferer()

    seg = inferer.apply(img_sitk)

    if not isinstance(seg, np.ndarray):
        seg = sitk.GetArrayFromImage(seg)

    # 🔥 Fallback if lungmask fails
    if np.sum(seg) == 0:
        print("⚠️ lungmask failed — using fallback segmentation")
        seg = (volume < np.percentile(volume, 60)).astype(np.uint8)

    return seg

print("Running lung segmentation...")
lung_mask_volume = lungmask_segment_safe(volume)

print("✅ Lung mask shape:", lung_mask_volume.shape)
print("✅ Lung voxels:", np.sum(lung_mask_volume))
```

```python
print("Min:", volume.min())
print("Max:", volume.max())
print("Mean:", volume.mean())

plt.hist(volume.flatten(), bins=50)
plt.title("Intensity Distribution")
plt.show()
```

```python
lung_pixels = volume[lung_mask_volume > 0]

if len(lung_pixels) == 0:
    print("⚠️ No lung pixels found — using global threshold")
    threshold = np.percentile(volume, 95)
else:
    threshold = np.percentile(lung_pixels, 99)

print("✅ Threshold:", threshold)
```

```python
def detect_tumor(img, threshold):
    return img > threshold
```

```python
def plot_slices_safe(volume, lung_masks, threshold, n=5):
    print("📊 Plotting slices...")

    total = volume.shape[0]
    idxs = np.linspace(0, total-1, n).astype(int)

    for idx in idxs:
        img = volume[idx]
        tumor = detect_tumor(img, threshold)

        plt.figure(figsize=(4,4))
        plt.imshow(img, cmap="gray")

        plt.imshow(np.ma.masked_where(~tumor, tumor),
                   cmap="Reds", alpha=0.5)

        plt.title(f"Slice {idx} | tumor pixels: {tumor.sum()}")
        plt.axis("off")
        plt.show()

plot_slices_safe(volume, lung_mask_volume, threshold)
```

```python
def compute_metrics(volume, threshold):
    total_volume = 0
    values = []

    for i in range(volume.shape[0]):
        tumor = detect_tumor(volume[i], threshold)

        total_volume += tumor.sum()

        if tumor.sum() > 0:
            values.extend(volume[i][tumor])

    if len(values) == 0:
        return {"Volume": 0, "Max": 0, "Mean": 0}

    return {
        "Volume": total_volume,
        "Max": np.max(values),
        "Mean": np.mean(values)
    }

summary = compute_metrics(volume, threshold)
print("✅ Tumor summary:", summary)
```

```python
summary_df = pd.DataFrame({
    "Patient": [patient_id],
    "Tumor_Volume_voxels": [summary["Volume"]],
    "Max_Intensity": [summary["Max"]],
    "Mean_Intensity": [summary["Mean"]]
})

print(summary_df)
```

```python
# Create output folders
os.makedirs("outputs/figures", exist_ok=True)
os.makedirs("outputs/tables", exist_ok=True)

# ---- Save metrics table ----
table_path = f"outputs/tables/{patient_id}_metrics.csv"
summary_df.to_csv(table_path, index=False)
print("📊 Table saved:", table_path)


# ---- Save visualization (combined figure) ----
def save_overlay_figure(volume, threshold, patient_id, n=5):
    total = volume.shape[0]
    idxs = np.linspace(0, total-1, n).astype(int)

    fig, axes = plt.subplots(1, n, figsize=(18,4))

    for i, idx in enumerate(idxs):
        img = volume[idx]
        tumor = detect_tumor(img, threshold)

        axes[i].imshow(img, cmap="gray")

        axes[i].imshow(
            np.ma.masked_where(~tumor, tumor),
            cmap="Reds", alpha=0.6
        )

        axes[i].set_title(f"Slice {idx}")
        axes[i].axis("off")

    plt.tight_layout()

    fig_path = f"outputs/figures/{patient_id}_overlay.png"
    plt.savefig(fig_path, dpi=200)
    print("🖼️ Figure saved:", fig_path)

    plt.close()


# Save figure
save_overlay_figure(volume, threshold, patient_id)
```
